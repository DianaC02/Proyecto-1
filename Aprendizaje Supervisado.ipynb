```python
#DESCARGAR Y PREPARAR LOS DATOS 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import roc_curve
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.utils import shuffle
import matplotlib.pyplot as plt


data= pd.read_csv('/datasets/Churn.csv')
data_copy = data.copy()
print(data.info())
```

```python
data_copy.drop(columns=['Surname', 'Geography'], inplace=True)
data_copy = data_copy[data_copy['Tenure'].notna()]
print(data_copy.info())
```

```python
#Codificar las columnas categóricas:
data_copy['Gender'] = pd.get_dummies(data_copy['Gender'], drop_first=True)
print(data_copy['Gender'])
```

<div class="alert alert-block alert-success">  
<b>Comentario del revisor</b> <a class="tocSkip"></a><br>  
<b>¡Gran trabajo!</b> 👏 - Excelente manera de preparar los datos iniciales. Me gusta cómo duplicaste el conjunto de datos original para mantener una copia intacta. Además, filtrar las filas con valores nulos y codificar las columnas categóricas son pasos fundamentales para un análisis limpio y preciso. ¡Sigue así, estás avanzando muy bien! 🚀  
</div>

```python
target=data_copy['Tenure']
features = data_copy.drop(['Tenure'], axis=1)
```

```python
# Separar los datos en entrenamiento y validación
features_train, features_valid, target_train, target_valid = train_test_split(
      features, target, test_size=0.35, random_state=12345)

# Lista de modelos con sus nombres y clases
models = [
    ("Linear Regression", LinearRegression()),
    ("Random Forest", RandomForestRegressor(n_estimators=200, random_state=1234)),
    ("Decision Tree", DecisionTreeRegressor(max_depth=25, random_state=1234))
]

# Iterar a través de los modelos y realizar el entrenamiento y evaluación
best_model = None
best_score = float('-inf')

for name, model in models:
    model.fit(features_train, target_train)
    probabilities_valid = model.predict(features_valid)
    score = model.score(features_valid, target_valid)

    print(f"{name}: R2 score = {score}")
    mae = mean_absolute_error(target_valid, probabilities_valid)
    print("MAE:", mae)

    # Si este modelo es mejor que el anterior mejor modelo, actualízalo
    if score > best_score:
        best_score = score
        best_model = model




    
```

# Un R² ligeramente positivo (como el de Linear Regression con 0.000968) indica que el modelo está explicando un poco más de la varianza que simplemente tomar la media de los valores. Por el contrario, un R² negativo en los otros modelos muestra que están rindiendo peor que la media.

```python
data.drop(columns=['Surname', 'Geography'], inplace=True)
```

```python
data['Gender'] = data['Gender'].apply(lambda x: 1 if x == 'Female' else 0)

missing_tenure = data[data['Tenure'].isna()].drop(columns=['Tenure'])
missing_tenure.fillna(missing_tenure.mean(), inplace=True)

# Predecir los valores faltantes usando el mejor modelo
predicted_tenure = best_model.predict(missing_tenure)

# Rellena los valores faltantes en el DataFrame original
data.loc[data['Tenure'].isna(), 'Tenure'] = predicted_tenure
```

```python
features=data.drop(['Exited'], axis=1)
target=data['Exited']
print(len(features), len(target))
print(data['Exited'].value_counts())
```

```python
# Separar los datos en entrenamiento y validación
features_train, features_valid, target_train, target_valid = train_test_split(
      features, target, test_size=0.25, random_state=12345)

# Lista de modelos con sus nombres y clases
models = [
    ("DecisionTreeClassifier", DecisionTreeClassifier()),
    ("RandomForestClassifier", RandomForestClassifier(n_estimators=200, random_state=1234)),
    ("LogisticRegression", LogisticRegression()
 )
]

# Iterar a través de los modelos y realizar el entrenamiento y evaluación
for name, model in models:
    # Entrenar el modelo con los datos de entrenamiento
    model.fit(features_train, target_train)
    
    #predicción
    probabilities_valid = model.predict(features_valid)
    
    # Evaluar el modelo en el conjunto de validación
    score = model.score(features_valid, target_valid)
    print('-----------------')
    print('NAME', name)
    print(confusion_matrix(target_valid,probabilities_valid))
    print(recall_score(target_valid, probabilities_valid))
    print(precision_score(target_valid, probabilities_valid))
    f1 = f1_score(target_valid,probabilities_valid)
    print(f1)
    print('-----------------')
    
    roc_curve(target_valid, probabilities_valid)

```

# RandomForestClassifier tiene la mayor precisión, pero sacrifica algo de recall.
# DecisionTreeClassifier está más equilibrado entre recall y precisión.
# LogisticRegression no logra predecir la clase minoritaria (clase 1) en absoluto. 
# Esto se debe a que las clase están desbalanceadas. 

<div class="alert alert-block alert-success">  
<b>Comentario del revisor</b> <a class="tocSkip"></a><br>  
<b>¡Buen trabajo!</b> 🎉 - Estás aplicando un enfoque integral al abordar el desbalanceo en los datos y evaluando múltiples modelos de clasificación con métricas clave como la precisión, el recall y el F1-score. Además, la comparación entre los modelos te ayudará a seleccionar el más adecuado para los datos. Muy bien identificar las limitaciones de la regresión logística en este caso. ¡Continúa experimentando y explorando formas de mejorar los resultados! 💪📊

```python
#balancear las clases mediante el submuestreo 
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_downsampled = pd.concat(
    [features_zeros.sample(frac=fraction, random_state=12345)]
    + [features_ones]
    )

    target_downsampled = pd.concat(
    [target_zeros.sample(frac=fraction, random_state=12345)]
    + [target_ones]
    )

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345
    )

    return features_downsampled, target_downsampled


features_downsampled, target_downsampled = downsample(
    features_train, target_train, 0.3
)



print(len(features_downsampled), len(target_downsampled))
print("Target Train \n",target_train.value_counts())
print("Target down \n",target_downsampled.value_counts())
```

```python
#Con el mejor modelo, cambiar los hiperparametros hasta que se ajuste al mejor 
model = DecisionTreeClassifier(
    max_depth=8, 
    min_samples_split=5, 
    min_samples_leaf=10, 
    max_features=6
)
model.fit(features_downsampled, target_downsampled)
probabilities_valid = model.predict(features_downsampled)
print("Recall:", recall_score(target_downsampled, probabilities_valid))
print("Precision:", precision_score(target_downsampled, probabilities_valid))
f1 = f1_score(target_downsampled, probabilities_valid)
print("F1 Score:", f1)
# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(target_downsampled, probabilities_valid)
# Graficar la curva ROC
plt.figure()
plt.plot(fpr, tpr, label='ROC curve')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--', label='Random guess')  # Línea diagonal
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend()
plt.show()

```

El nuevo modelo mejoró en todas las métricas:

Recall (0.734): La capacidad del modelo para identificar correctamente las clases positivas ha aumentado, lo que significa que está capturando más de los verdaderos positivos.
Precision (0.804): El modelo es más preciso al predecir clases positivas, lo que indica que tiene menos falsos positivos.
F1 Score (0.768): Como combinación del recall y la precisión, esta métrica sugiere un mejor equilibrio entre la sensibilidad y la precisión del modelo.
En esta gráfica, la curva ROC está por encima de la línea aleatoria, lo que sugiere que el modelo tiene un desempeño aceptable para distinguir entre clases.
