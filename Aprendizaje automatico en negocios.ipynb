```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Cargar los datos y verificar las primeras 5 columnas 
data_region_0= pd.read_csv('/datasets/geo_data_0.csv')
data_region_1= pd.read_csv('/datasets/geo_data_1.csv')
data_region_2= pd.read_csv('/datasets/geo_data_2.csv')
print("Datos de la Región 0:")
print(data_region_0.head())
print("\nDatos de la Región 1:")
print(data_region_1.head())
print("\nDatos de la Región 2:")
print(data_region_2.head())
```

```python
#Verificar datos faltantes y tipo de dato 
print("Resumen de la Región 0:")
print(data_region_0.info())
print("\nResumen de la Región 1:")
print(data_region_1.info())
print("\nResumen de la Región 2:")
print(data_region_2.info())
```

```python
print(data_region_0.describe())
print(data_region_1.describe())
print(data_region_2.describe())

```

```python
#Verificar que no hayan valores faltantes 
print(data_region_0.isna())
print(data_region_1.isna())
print(data_region_2.isna())
```

```python
# Definir objetivo y características
def separar_caracteristicas_objetivo(data):
    features = data.drop(['product', 'id'], axis=1)
    target = data['product']
    return features, target

# Dividir los datos en conjunto de entrenamiento y validación
def dividir_datos(features, target, test_size=0.25, random_state=12345):
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=test_size, random_state=random_state)
    return features_train, features_valid, target_train, target_valid

# Entrenar y evaluar el modelo
def entrenar_y_evaluar(features_train, features_valid, target_train, target_valid):
    model = LinearRegression()
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    mse = mean_squared_error(target_valid, predictions_valid)
    rmse = mse ** 0.5
    print(f'RMSE: {rmse:.2f}')
    print(f'Predicción media: {predictions_valid.mean():.2f}')
    print(f'Valor real medio: {target_valid.mean():.2f}\n')
    return model, predictions_valid

# Función para procesar cada región
def procesar_region(data_region, nombre_region):
    features, target = separar_caracteristicas_objetivo(data_region)
    features_train, features_valid, target_train, target_valid = dividir_datos(features, target)
    model, predictions_valid = entrenar_y_evaluar(features_train, features_valid, target_train, target_valid)
    print("predition valid count", len(predictions_valid))
    top_10_predictions = sorted(predictions_valid, reverse=True)[:10]
    print(f"Las 10 predicciones más altas para la {nombre_region} son:", top_10_predictions)
    return target_valid, predictions_valid
# Aplicar la función a cada región
target_valid, predictions_valid = procesar_region(data_region_0, "región 0")
procesar_region(data_region_1, "región 1")
procesar_region(data_region_2, "región 2")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_top_10_predicciones(target_valid, predictions_valid, nombre_region):
    plt.figure(figsize=(8, 5))

    # Convertir a numpy arrays (para evitar problemas con índices de Pandas)
    target_valid = np.array(target_valid)
    predictions_valid = np.array(predictions_valid)

    # Obtener los índices de los 10 valores reales más altos
    top_10_indices = np.argsort(target_valid)[-10:]  # Últimos 10 valores más altos

    # Extraer los valores reales y predicciones correspondientes
    top_10_real = target_valid[top_10_indices]
    top_10_pred = predictions_valid[top_10_indices]

    # Graficar los valores reales y predichos
    plt.bar(range(10), top_10_real, label="Valores Reales", alpha=0.7)
    plt.bar(range(10), top_10_pred, label="Predicciones", alpha=0.7)

    plt.xticks(range(10), labels=[f'Dato {i+1}' for i in range(10)])
    plt.xlabel('Top 10 Valores Reales Más Altos')
    plt.ylabel('Valor')
    plt.title(f'Comparación de Predicciones en los Valores Más Altos - {nombre_region}')
    plt.legend()
    
    plt.show()

# Llamar a la función para una región como prueba
graficar_top_10_predicciones(target_valid, predictions_valid, "Región 0")


```

```python
# Definir variables clave
inversion_total = 100_000_000  # Inversión total en dólares
numero_pozos = 200  # Número de pozos
ingreso_por_unidad_de_producto = 4500  # Ingreso por unidad de producto en dólares

# Calcular la inversión por pozo
inversion_por_pozo = inversion_total / numero_pozos  # Inversión por pozo en dólares

# Calcular la producción mínima por pozo en miles de barriles para evitar pérdidas
produccion_minima_por_pozo = inversion_por_pozo / ingreso_por_unidad_de_producto  # Unidades necesarias
print(f"Producción mínima requerida por pozo: {produccion_minima_por_pozo:.2f} unidades (en miles de barriles)")


# Valores reales medios de reservas por pozo en cada región (en miles de barriles)
regiones = {
    "Región 0": 92.08,
    "Región 1": 68.72,
    "Región 2": 94.88
}

# Comparar cada región directamente con el umbral y mostrar el resultado
for nombre, valor_medio in regiones.items():
    diferencia = valor_medio - produccion_minima_por_pozo
    resultado = "Cumple" if diferencia >= 0 else "No cumple"
    print(f"{nombre}: diferencia = {diferencia:.2f} -> {resultado}")
```

```python
# --- Función nueva para extraer los top 200 pozos reales ---
def get_top_n_predictions(data_region, model, n_sample=500, top_n=200):
    """
    Extrae una muestra de n_sample pozos de la región, predice sus reservas
    usando el modelo entrenado y retorna los top_n pozos con mayor predicción.
    """
    # Tomar una muestra aleatoria de n_sample pozos
    sample_data = data_region.sample(n=n_sample, random_state=12345)
    features_sample, _ = separar_caracteristicas_objetivo(sample_data)
    predictions = model.predict(features_sample)
    # Ordenar y seleccionar los índices de los top_n pozos
    indices_top = np.argsort(predictions)[::-1][:top_n]
    top_predictions = predictions[indices_top]
    return top_predictions
```

```python
def calcular_ganancia(predicciones, inversion_total=100_000_000, ingreso_por_unidad=4500, n_pozos=200):
    """
    Calcula la ganancia potencial de los n_pozos con mayores predicciones.

    Parámetros:
    - predicciones: lista o array de predicciones de reservas (en miles de barriles).
    - inversion_total: inversión total en dólares (default: 100 millones).
    - ingreso_por_unidad: ingreso en dólares por cada unidad (mil barriles) (default: 4500).
    - n_pozos: cantidad de pozos a seleccionar (default: 200).

    Retorna:
    - predicciones_top: array con las predicciones de los n_pozos seleccionados.
    - total_reservas: suma del volumen de reservas de los pozos seleccionados (en miles de barriles).
    - ingreso_total: ingreso total potencial en dólares.
    - ganancia: ganancia potencial (ingreso_total - inversion_total).
    """
    # Convertir a array en caso de que no lo sea
    predicciones = np.array(predicciones)
    
    # Ordenar los índices de mayor a menor y seleccionar los primeros n_pozos
    indices_top = np.argsort(predicciones)[::-1][:n_pozos]
    predicciones_top = predicciones[indices_top]
    
    # Sumar el volumen de reservas de los pozos seleccionados
    total_reservas = predicciones_top.sum()
    
    # Calcular el ingreso total: cada unidad (mil barriles) genera 4500 USD
    ingreso_total = total_reservas * ingreso_por_unidad
    
    # Calcular la ganancia potencial
    ganancia = ingreso_total - inversion_total
    
    return predicciones_top, total_reservas, ingreso_total, ganancia

# Extraer los 200 pozos con mayor predicción para cada región usando datos reales
# Entrenar modelo para Región 0
features_0, target_0 = separar_caracteristicas_objetivo(data_region_0)
features_train_0, features_valid_0, target_train_0, target_valid_0 = dividir_datos(features_0, target_0)
model_0, _ = entrenar_y_evaluar(features_train_0, features_valid_0, target_train_0, target_valid_0)

# Entrenar modelo para Región 1
features_1, target_1 = separar_caracteristicas_objetivo(data_region_1)
features_train_1, features_valid_1, target_train_1, target_valid_1 = dividir_datos(features_1, target_1)
model_1, _ = entrenar_y_evaluar(features_train_1, features_valid_1, target_train_1, target_valid_1)

# Entrenar modelo para Región 2
features_2, target_2 = separar_caracteristicas_objetivo(data_region_2)
features_train_2, features_valid_2, target_train_2, target_valid_2 = dividir_datos(features_2, target_2)
model_2, _ = entrenar_y_evaluar(features_train_2, features_valid_2, target_train_2, target_valid_2)

top_predictions_0 = get_top_n_predictions(data_region_0, model_0, n_sample=500, top_n=200)
top_predictions_1 = get_top_n_predictions(data_region_1, model_1, n_sample=500, top_n=200)
top_predictions_2 = get_top_n_predictions(data_region_2, model_2, n_sample=500, top_n=200)

# Calcular la ganancia potencial para cada región
resultados = {}
for i, top_predictions in enumerate([top_predictions_0, top_predictions_1, top_predictions_2]):
    pred_top, total_reservas, ingreso_total, ganancia = calcular_ganancia(top_predictions)
    resultados[f"Región {i}"] = {
        "predicciones_top": pred_top,
        "total_reservas": total_reservas,
        "ingreso_total": ingreso_total,
        "ganancia": ganancia
    }
    print(f"Región {i}:")
    print(f"  Total reservas: {total_reservas:.2f} miles de barriles")
    print(f"  Ingreso total: ${ingreso_total:,.2f}")
    print(f"  Ganancia potencial: ${ganancia:,.2f}\n")

# Conclusión:
# Se seleccionará la región con la ganancia potencial más alta.
region_seleccionada = max(resultados.items(), key=lambda x: x[1]['ganancia'])
print("La región seleccionada es:", region_seleccionada[0])
```

```python
def bootstrapping_beneficios(predicciones_top, inversion_total=100_000_000, ingreso_por_unidad=4500, n_samples=1000):
    """
    Emplea bootstrapping para estimar la distribución de beneficios de un conjunto de pozos.

    Parámetros:
    - predicciones_top: array de predicciones (en miles de barriles) de los 200 pozos seleccionados.
    - inversion_total: inversión total en dólares (default: 100 millones).
    - ingreso_por_unidad: ingreso en dólares por cada unidad (mil barriles) (default: 4500).
    - n_samples: número de muestras bootstrap (default: 1000).

    Retorna:
    - beneficios: array con la ganancia potencial calculada en cada muestra.
    - beneficio_promedio: media de los beneficios.
    - ci_95: intervalo de confianza del 95% (percentiles 2.5 y 97.5).
    - riesgo_perdida: porcentaje de muestras donde la ganancia es negativa.
    """
    n_pozos = len(predicciones_top)
    beneficios = np.empty(n_samples)
    
    for i in range(n_samples):
        # Muestra con reemplazo de los 200 pozos
        muestra = np.random.choice(predicciones_top, size=n_pozos, replace=True)
        # Sumar reservas de la muestra
        total_reservas = muestra.sum()
        # Calcular ingreso total potencial
        ingreso_total = total_reservas * ingreso_por_unidad
        # Calcular beneficio (ganancia) restando la inversión total
        beneficios[i] = ingreso_total - inversion_total
    
    beneficio_promedio = beneficios.mean()
    ci_95 = np.percentile(beneficios, [2.5, 97.5])
    riesgo_perdida = (beneficios < 0).mean() * 100  
    
    return beneficios, beneficio_promedio, ci_95, riesgo_perdida



# Calcular bootstrapping para cada región utilizando las predicciones reales
resultados_bootstrap = {}
for i, top_predictions in enumerate([top_predictions_0, top_predictions_1, top_predictions_2]):
    beneficios, beneficio_promedio, ci_95, riesgo_perdida = bootstrapping_beneficios(top_predictions)
    resultados_bootstrap[f"Región {i}"] = {
        "beneficio_promedio": beneficio_promedio,
        "ci_95": ci_95,
        "riesgo_perdida": riesgo_perdida
    }
    print(f"Región {i}:")
    print(f"  Beneficio promedio: ${beneficio_promedio:,.2f}")
    print(f"  Intervalo de confianza 95%: ${ci_95[0]:,.2f} a ${ci_95[1]:,.2f}")
    print(f"  Riesgo de pérdidas: {riesgo_perdida:.2f}%\n")

# Seleccionar la región con el beneficio promedio más alto (y, por supuesto, revisar el riesgo de pérdidas)
region_seleccionada = max(resultados_bootstrap.items(), key=lambda x: x[1]['beneficio_promedio'])
print("La región seleccionada mediante bootstrapping es:", region_seleccionada[0])
```

Conclusión del análisis:

Riesgo de pérdidas:
Todas las regiones presentan un riesgo inferior al 2.5%, por lo que son viables desde el punto de vista de riesgo.

Beneficio promedio:
La Región 0 muestra el beneficio promedio más alto ($3,496,811.69), lo que indica que, en promedio, la inversión generaría mayores ganancias.

Intervalo de confianza:
Aunque los intervalos de confianza se solapan en cierta medida, el hecho de que la Región 0 tenga el beneficio promedio superior y un riesgo de pérdidas nulo refuerza la elección.
